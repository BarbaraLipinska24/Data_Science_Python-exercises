```python
import sklearn
```

Zainicjowanie danych

```python
import numpy as np

x=np.arange(0,50,0.5)
noise=10*np.random.randn(100)
y=2*x+100+noise
x=x.reshape(-1,1)

print('Rozmiar X:', x.shape)
print('Rozmiar y:', y.shape)
```

Podział danych na testowe i treningowe

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25)

```

Wizualizacja


```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()

plt.figure(figsize=(7,5))
plt.scatter(x_train,y_train,c='b',label='zbiór treningowy')
plt.scatter(x_test,y_test,c='r',label='zbiór testowy')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
```

Regresja liniowa

```python
from sklearn.linear_model import LinearRegression

lin_reg=LinearRegression()
lin_reg.fit(x_train,y_train)

print(lin_reg.coef_) #a
print(lin_reg.intercept_) #wyraz wolny
```

Wizualizacja zbioru treningowego i dopasowanego modelu

```python
plt.figure(figsize=(7,5))
plt.title('Regresja Liniowa - jedna zmienna')
plt.scatter(x_train,y_train,c='b',label='Zbiór treningowy')
plt.plot(x,lin_reg.intercept_+lin_reg.coef_[0]*x,c='r')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
```

Wizualizacja zbioru testowego i dopasowania

```python
plt.figure(figsize=(7,5))
plt.title('Regresja Liniowa - jedna zmienna')
plt.scatter(x_test,y_test,c='b',label='Zbiór treningowy')
plt.plot(x,lin_reg.intercept_+lin_reg.coef_[0]*x,c='r')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
```

Ocena modelu

```python
lin_reg.score(x_test,y_test)
```
